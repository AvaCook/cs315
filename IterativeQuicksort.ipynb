```python
"""
Name: Ava Cook
Date: February 2026
Purpose: Iterative implementation of Quicksort
         using a stack instead of recursion.

Sources:
- https://stackoverflow.com/questions/12553238/quicksort-iterative-or-recursive
- my old code from Algorithms and Analysis
- Chatgpt for errors/small edits
"""

def Partition(A, p, r):
    x = A[r]
    i = p - 1

    for j in range(p, r):
        if A[j] <= x:
            i += 1
            A[i], A[j] = A[j], A[i]

    A[i + 1], A[r] = A[r], A[i + 1]
    q = i + 1

    print("after Partition, p,q,r,A:", p, q, r, A)

    return q

def IterativeQuicksort(A):
    stack = []

    stack.append((0, len(A) - 1))

    while stack:
        p, r = stack.pop()

        if p < r:
            q = Partition(A, p, r)

            if q + 1 < r:
                stack.append((q + 1, r))

            if p < q - 1:
                stack.append((p, q - 1))

# test
A = [3, 0, 8, 2, 5, 8, 8, 2]

print("initial array:", A)
IterativeQuicksort(A)
print("sorted array:", A)
```

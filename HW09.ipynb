```python
import math
import sys


def caesar_square_decrypt(cipher: str) -> str:
	L = len(cipher)
	if L == 0:
		return ""

	cols = math.ceil(math.sqrt(L))
	rows = math.ceil(L / cols)

	# Create empty rows x cols matrix
	matrix = [[''] * cols for _ in range(rows)]

	# Fill column by column with ciphertext characters
	idx = 0
	for c in range(cols):
		for r in range(rows):
			if idx < L:
				matrix[r][c] = cipher[idx]
				idx += 1

	# Read row by row to reconstruct plaintext
	plaintext_chars = []
	for r in range(rows):
		for c in range(cols):
			ch = matrix[r][c]
			if ch:
				plaintext_chars.append(ch)

	return ''.join(plaintext_chars)


print(caesar_square_decrypt('NARRFSECCNTCXAIOUITEPSSIISHHEOSAEIAK'))
```

```python
"""
Name: Ava Cook
Date: February 2026
Purpose: Recursive implementation of Quicksort that prints the array
         after each call to Partition.

Sources:
- https://en.wikipedia.org/wiki/Quicksort
- my old code from Algorithms and Analysis
- Chatgpt for errors/small edits
"""

def Partition(A, p, r):
    x = A[r]
    i = p - 1

    for j in range(p, r):
        if A[j] <= x:
            i += 1
            A[i], A[j] = A[j], A[i]

    A[i + 1], A[r] = A[r], A[i + 1]
    q = i + 1

    print("after Partition, p,q,r,A:", p, q, r, A)

    return q


def Quicksort(A, p, r):
    if p < r:
        q = Partition(A, p, r)
        Quicksort(A, p, q - 1)
        Quicksort(A, q + 1, r)


# test
A = [3, 0, 8, 2, 5, 8, 8, 2]

print("initial array:", A)
Quicksort(A, 0, len(A) - 1)
print("sorted array:", A)
```
